```python
import seaborn as sns
```

## Getting Dataset using seaborn

```python
cars_mpg = sns.load_dataset('mpg')
usa_cars_mpg=cars_mpg[cars_mpg['origin']=='usa']
europe_cars_mpg=cars_mpg[cars_mpg['origin']=='europe']
japan_cars_mpg=cars_mpg[cars_mpg['origin']=='japan']
cars_mpg
```

## Give theme for tha graph

```python
sns.set()
```

## Graph model (MPG vs Horsepower) for cars made in USA

```python
sns.relplot(x='horsepower', y='mpg', kind='line',color = 'brown', data = usa_cars_mpg)
```

## Graph model (MPG vs Horsepower) for cars made in Japan

```python
sns.relplot(x='horsepower', y='mpg', kind='line',color = 'green', data = japan_cars_mpg)
```

## Graph model (MPG vs Horsepower) for cars made in Europe

```python
sns.relplot(x='horsepower', y='mpg', kind='line',color = 'yellow', data = europe_cars_mpg)
```

## Conclusions

1- As you can see from the graphs the USA cars have higher horse power and it effect it MPG to be lower.

2- As you can see from the graphs the Japan cars have fuel efficiency cars with less horse power and it effect it MPG to be higher.

3- As you can see from the graphs the Europe cars have some balence in their cars with less horse power and it effect it MPG to be good efficienet.

4- Europe cars has the highest mpg and the lowest horsepower.

5- USA cars has the highest horsepower and the least MPG.
